# **How much more do organic avocado costs?**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("avocado.csv").drop("Unnamed: 0", axis=1)
conventional = df[(df["type"]=="conventional") & (df["AveragePrice"] != np.nan)]["AveragePrice"].values
organic = df[(df["type"]=="organic") & (df["AveragePrice"] != np.nan)]["AveragePrice"].values

# Calculate and print the average price conventional avocados
conventional_mean = np.mean(conventional)
print('Mean conventional price: ${:.2f}'.format(conventional_mean))

# Calculate the monthly premium for organic avocados
conventional = conventional[:9123]
premium = (organic - conventional) / conventional

# Calculate the minimum and maximum premium
premium_min = np.min(premium)
premium_max = np.max(premium)

# Print the range of the premium 
msg = 'Range of premium: {:.1%} to {:.1%}'
print(msg.format(premium_min,premium_max))
```

### **Making NumPy arrays from lists**

* Lists can even contain other lists. 
 *  Create a list `weekdays` where each element is a daily list, starting on Monday.
 * Create a NumPy array `traffic` from the list `weekdays`.
 * Calculate the mean hourly traffic.
 * Plot the mean hourly traffic for weekdays.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


monday, tuesday, wednesday, thursday, friday, saturday, sunday = [], [], [], [], [], [], []
df = pd.read_csv("fremont_bridge_hourly_weekdays.csv").drop("Hour", axis=1)
for i, j in zip(df.columns, [monday, tuesday, wednesday, thursday, friday, saturday, sunday]):
    j.extend(list(df[i].values))

# Create a list "weekdays" where each element is a daily list, starting on Monday
weekdays = [monday, tuesday, wednesday, thursday, friday]

# Create a NumPy array "traffic" from the list "weekdays"
traffic = np.array(weekdays)

# Calculate the hourly mean traffic
hourly = np.mean(traffic,axis=0)

# Plot the mean hourly traffic for weekdays
plt.plot(hourly)
plt.show()
```

### **Operating on lists and arrays**

* When two NumPy arrays are added together with `+`, Python adds the two arrays together element-by-element and returns a new array of the same shape as the original arrays. 
* This is similar to how MATLAB handles addition with matrices.
* On the other hand, adding two lists together with `+` in Python concatenates them.


```python
from numpy import array
# Create two lists, x and y, containing the 3 integers in each
x = [1, 2, 3]
y = [4, 5, 6]

# Add them together & save the result to z
z = x + y
print(z)
print(len(z) == len(x)+len(y))

x = array(x)
y = array(y)
z = x + y
print(z)
print(z.shape == x.shape)
```

### **Encoding data in color & size**

### **Determine engine types in wildlife strikes**
* The list `engine_types_labels` contains descriptors of the five different types of engines.
The NumPy array `engine_types` contains the integers `0` through `4`, specifying the type of engine, where the integer corresponds to the index of the `engine_types_labels` list.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("FAA_wildlife_strikes_CA.csv").drop("Unnamed: 0", axis=1)
df["TYPE_ENG"] = df["TYPE_ENG"].astype(str)
df = df[df["TYPE_ENG"]!="nan"]
engine_types_labels = ['Reciprocating', 'Turbofan', 'Turbojet', 'Turboprop', 'Turboshaft (helicopter)']
engine_dict = {'Reciprocating':0, 'Turbofan':1, 'Turbojet':2, 'Turboprop':3, 'Turboshaft (helicopter)':4}
engine_types = np.array(list(map(engine_dict.get, df["TYPE_ENG"].values)))

# Create the histogram with one bin for each unique value in engine_types
plt.figure()
plt.hist(engine_types,bins=[0,1,2,3,4,5])

# Create x-ticks with the labels in engine_types_labels
plt.xticks([0.5,1.5,2.5,3.5,4.5],
            engine_types_labels, 
            rotation=30)

# Show the plot
plt.show()
```

### **Counting different types of aircraft**

* Write a for loop that goes through each key:value pair of `airplane_types`.
* On each iteration, "The x airplane has y engines." should be printed out, where x is the name of the airplane and y is the type of engine.
* For each iteration, increment the values of the engine_type key in the `counts` dictionary.
* Finally, print the "`counts`" dictionary

```python
airplane_types = {'A-319': 'Turbofan', 'A-320': 'Turbofan', 'B-737': 'Turbofan', 'B-737-300': 'Turbofan', 'B-737-500': 'Turbofan', 'B-737-700': 'Turbofan', 'B-737-800': 'Turbofan', 'B-757-200': 'Turbofan', 'B-767-300': 'Turbofan', 'C-172': 'Reciprocating', 'CRJ100/200': 'Turbofan', 'CRJ700': 'Turbofan', 'EMB-120': 'Turboprop', 'MD-80': 'Turbofan'}
counts = {}
# Loop over the key:value pairs in airplane_types
for airplane, engine_type in airplane_types.items():
    # Print the aircraft name and engine type of each aircraft
    print("The {} airplane has {} engines.".format(airplane, engine_type))
    # Increment the values of the engine_type key in the counts dictionary
    if engine_type in counts:
        counts[engine_type] = counts[engine_type] + 1
    else:
        counts[engine_type] = 1

print(counts)
  
```

### **Boolean indexing**

* Create a Boolean index `mammals` by finding records where the "Class" is "Mammalia".
* Create a Boolean index `birds` by finding records where the "Class" is "Aves".
* Use `numpy` to find the mean of the "Litter/Clutch size" column for mammals and birds.

```python
# Create Boolean indices for mammals and birds
import pandas as pd
import numpy as np
animals = pd.read_csv("animal.csv")

mammals = animals['Class']=='Mammalia'
birds = animals['Class']=='Aves'

# Use numpy and the Boolean indices to determine mean Litter/Clutch size
litter = np.mean(animals[mammals]['Litter/Clutch size'])
clutch = np.mean(animals[birds]['Litter/Clutch size'])

# Print the average Litter/Clutch size of each class
print('Mammals have an average of {} offspring in each litter.'.format(round(litter, 2)))
print('The average clutch size in a single brood is {} eggs.'.format(round(clutch, 2)))
```

### **Boolean indexing and Matplotlib fun**

* Create a Boolean filter for '`Turbofan`' in the '`Engine`' column.
Create a Boolean filter for `'Turboprop'` in the `'Engine'` column.
Plot two scatter plots using `turbofan` and `turboprop` to filter the `strikes` dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt


strikes = pd.read_csv("FAA_wildlife_strikes_CA.csv").drop("Unnamed: 0", axis=1)[["INCIDENT_DATE",  "SPEED", "HEIGHT", "TYPE_ENG"]].dropna()
strikes.columns = ["Date",  "Speed",   "Height", "Engine"]
strikes = strikes.drop_duplicates().reset_index(drop=True)


# Filter strikes by engine type
turbofan = strikes['Engine']=='Turbofan'
turboprop = strikes['Engine']=='Turboprop'

# Create scatter plot of speed and height for each engine type
plt.scatter(strikes[turbofan]['Speed'],strikes[turbofan]['Height'],label='Turbofan')
plt.scatter(strikes[turboprop]['Speed'],strikes[turboprop]['Height'],label='Turboprop')
plt.legend()
plt.xlabel('Strike speed (knots)')
plt.ylabel('Strike height (feet)')
plt.show()
```
